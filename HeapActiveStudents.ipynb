```python
keys = ["blog"]
blog = heap[heap['initial_landing_page'].str.contains('|'.join(keys)).fillna(False)]
```

```python
blog
```

```python
# of enrollments from the blog 
len(blog)
```

```python
#enrollments from the blog, broken down by initial referring page
blog.initial_landing_page.value_counts()
```

```python
#grouping workshops and landing page segments
blog.groupby(['workshop_shortname', 'initial_landing_page']).user_id.count()
```

```python
#Number of blog enrollments by questionnaire date (close proxy to enrollment date)
blog.questionnaire_deadline.value_counts()
```

```python
#number of blog enrollments broken down by enrollment type and enrollment time
blog.groupby(['questionnaire_deadline', 'workshop_shortname']).user_id.count()
```
