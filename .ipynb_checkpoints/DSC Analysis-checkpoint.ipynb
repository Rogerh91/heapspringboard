```python
typeform = typeform.rename(columns = {"Submit Date (UTC)" : "submit_date"})
```

```python
cols = typeform.columns.tolist()
cols.insert(0, cols.pop(cols.index('submit_date')))
typeform = typeform.reindex(columns = cols)
typeform.set_index("submit_date")
```

```python
typeform = typeform.rename(columns = {"Submit Date (UTC)" : "submit_date", "How did you hear about the Data Science Career Track?" : "attribution"})
```

```python
typefinal = typeform[(typeform["submit_date"] <= '2017-01-16')]
```

```python
len(typefinal)
```

```python
typefinal.attribution.value_counts()
```

```python
typefinal = typefinal.fillna(0)
typefinal = typefinal.rename(columns = {"Email" :  "email"})
typefinal
```

```python
typeheap = pd.merge(typefinal,heap, on = "email")
```
